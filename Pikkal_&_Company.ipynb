<h2>Simulating the survey for 20000 people as specified in the report<h2>
<p>A dataframe is created with columns id(signifying 20000 people), stress(the stress acceptance levels),safety(corresponding to the question if the person is feeling safe when returning to their physical workplace),org future(whether the person is confident about the safety of the organization where he/she works). The first three columns has values from 0 to 4  where 0 is 'strongly agree', 1 is 'agree' and so on. The columns wfh corresponds to the readyness of how much they are interested in working from home. Here 0 means '100%' ready to work and 4 represents 'Never'. The last column is the represenation of the industry to which the person belongs to. The numerical values are a representation of their respective position in the wfh_sector dataframe, where 0 is 'Consumer Industry', 1 is 'Education' and so on.<p>
<p>
All the columns apart from id column eis populated in accordance with the probability of the values as per the report. As an example, 'stress is acceptable' statement is strongly agreed by 14% of the total survey respodents. This is incorporated into the values assigned to the columns of the dataframe using the np.random.choice() function with specified probabilities of occurance of these responses during the survey.  


```
survey = pd.DataFrame(columns=['id','stress','safety','org future','wfh','industry'])
for i in range(20000):
  survey.loc[i] = [i,int(np.random.choice(5,1,p=[.14,.31,.31,.19,.05])),int(np.random.choice(5,1,p=[.13,.25,.34,.19,.09])),int(np.random.choice(5,1,p=[.44,.36,.16,.03,.01])),int(np.random.choice(5,1,p=[.17,.34,.32,.09,.08])),int(np.random.choice(7,1,p=[.19,.04,.11,.07,.06,.30,.23]))]
survey.head()
```

<h2><center>Visualizations</center><h2>

<p> Creating a horizontal bar chart to visualize the responses to different questions

```
sns.set_style('white')
fig, ax = plt.subplots()
poll1.plot(kind='barh', x='category',ax=ax)
ax.legend(loc='lower right')
plt.title('Survey Responses')
plt.show()
```

<p>The bar chart is vague and a little bit difficult to understand. So let's create a line chart for the respective questions.</p>

```
trans = pd.DataFrame(poll1.T[1:])
trans.columns=poll1.T.iloc[0]
fig, ax = plt.subplots()
trans.plot(marker='o', ax=ax)
ax.legend(loc='upper right')
fig.set_figwidth(10)
plt.title('Survey Responses')
plt.show()
```

<p> We can clearly see a correlation between the stress acceptance levels and the confidence in safety for returning to work. While those who strongly agree with the respective statements are quite low, most people are optimistic about the same. Another noticing trend in the three categories (stress,stress with depedent, safety fears) is the low number of people who are really opposed to the ideas presented during the survey. And we can also see that people with dependants are more stressed that those without. This is expected.</p>
<p> The main concept we can grasp from this graph is that majority of people working for different organizations are highly optimistic about the future of their respective companies.</p>

```
fig, ax = plt.subplots(1,3)
stress_sector.plot(kind='bar', x='sector', ax=ax[0], color=['red','green'])
stress_company_size.plot(kind='bar', x='Company Size', ax=ax[1], color=['red','green'])
stress_age.plot(kind='bar', x='age', ax=ax[2], color=['red','green'])
fig.set_figwidth(20)
fig.suptitle("Stress Level Acceptance")
plt.show()
```

Majority of people working in all sectors agree that the stress is acceptable, this is true irrespective of  different sectors or the size of the company or the age group.

```
trans2 = pd.DataFrame(wfh_sector.T[1:])
trans2.columns = wfh_sector.T.iloc[0]
trans3 = pd.DataFrame(wfh_company_size.T[1:])
trans3.columns = wfh_company_size.T.iloc[0]
fig, ax = plt.subplots(1,2)
trans2.plot(marker='o', ax=ax[0])
trans3.plot(marker='o', ax=ax[1])
fig.set_figwidth(20)
fig.suptitle('Confidence in future of company')
plt.show()
```

While there are people in certain sectors(mainly education and industrial, but if you look closely people in healthcare too.)who are against the idea </p>
<p> Although we can see same pattern of favoring work from home in companies of all sizes, we can easily see that a significant portion of respondents from small companies was against the idea completely. This maybe due to high interpersonal relations present in small companies.
